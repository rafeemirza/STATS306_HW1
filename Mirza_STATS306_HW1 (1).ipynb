Problem 2 is based on the `iris` data frame with 150 rows and 5 columns. The first 6 rows have been displayed in the following cell. You can run `help(iris)` to get the description of the data set.

````r
head(iris)
````

**(a)** Following the example in the lab, reproduce the following density plot for petal length with each iris species. What insights can you derive by looking at this density plot?: _1 points_

p2a.png

````r
ggplot (
      data=iris,
      aes(x = Petal.Length, color = Species, fill = Species)) + geom_density(alpha=0.5, linewidth = 0.4, color = "black") + labs(title = "Density Plot of Petal Length Across Species", x = "Petal Length (cm)", y = "Density")
````

**(b)** How are petal length and petal width related? Following the example we saw in lecture, reproduce the following plot: _1 points_


p2b.png

````r
ggplot() +
  geom_point(
    data = iris,
    mapping = aes(x = Petal.Length, y = Petal.Width, color = Species, shape = Species)
  ) +
  geom_smooth(
    data = iris,
    mapping = aes(x = Petal.Length, y = Petal.Width)
  ) + labs(title = 'Relation Between Petal Length and Petal Width for Iris Flowers', x = 'Petal Length (cm)', y = 'Petal Width (cm)')
````

**(c)**
The preceding plot indicates that the petal width is roughly linearly related to the petal length. Check this by computing the correlation of petal width  with petal length, and round the correlation to two decimal places.
````r
question2c <- cor(iris$Petal.Length, iris$Petal.Width)
round(question2c, 2)
````

## Problem 3: Starwars (3 pts)
This problem looks at the `starwars` dataset, which is part of dplyr package. Run the following code to get only those rows that do not have NA values for height and species (so we cleaned it a bit)

````r
library(dplyr)
starwars <- filter(starwars, !is.na(starwars$height), !is.na(starwars$species))
````

````r
glimpse(starwars)
````

**(a)** Use `filter()` to subset the starwars table down to only those characters from `homeworld` Naboo and Tatooine.  _1/2 point_

````r
starwars1 <- starwars %>%
              filter(homeworld == "Tatooine" | homeworld == "Naboo")
````

**(b)** Using the table you created in part (a), create the following plot showing the box and whisker plot of "height" for different "species" within each "homeworld.": _1 point_

**Hint:** Just like the 'color' attribute, there is a 'fill' attribute as well that can be used in aesthetics

download.png

````r
library(tidyverse)

ggplot(starwars1, mapping = aes(x = species, y = height, fill = homeworld)) + geom_boxplot(linewidth = 0.8, color = "black")
````

**(c)** Use `filter` to subset the starwars table down to only `mass` smaller than 1000. _1/2 point_

````r
starwars2 <- starwars %>%
            filter(mass < 1000)
````

**(d)** Using the table you created in part (c), recreate the following plot showing the relationship between mass and height. Add a smoothed linear regression line: _1 point_

download-1.png
